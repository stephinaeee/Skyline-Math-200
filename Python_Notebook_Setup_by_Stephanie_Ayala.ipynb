Research Question #1 :  Is there an association between how many close friends can you confide in and the thought of committing suicide?

Codebook 1

Screenshot 2023-04-30 at 3.33.56 PM.png

```python
#Manage data for variable 1
df['H4WS4'].value_counts()
```

```python
df['Close_Friends'] = df['H4WS4']
df['Close_Friends'].value_counts()
```

```python
df['Close_Friends'] = df['Close_Friends'].loc[ df['Close_Friends'] != ' ']
df['Close_Friends'].value_counts()
```

```python
df['Close_Friends'] = df['Close_Friends'].loc[ df['Close_Friends'] != '95']
df['Close_Friends'] = df['Close_Friends'].loc[ df['Close_Friends'] != '96']
df['Close_Friends'] = df['Close_Friends'].loc[ df['Close_Friends'] != '98']
df['Close_Friends'].value_counts()
```

```python
df['Close_Friends'].replace(['1','2','3','4','5'],['NONE','1-2 FRIENDS','3-5 FRIENDS','6-9 FRIENDS','10 OR MORE FRIENDS'], inplace = True)
df['Close_Friends'].value_counts()
```

```python
rp.summary_cat(df['Close_Friends'])
```

```python
#Graphs and tables or statistics
explode = (0.01, 0.01, 0.01, 0.2, 0.3)
df['Close_Friends'].value_counts().plot(kind ='pie', autopct='%1.1f%%', colors=['beige', 'rosybrown', 'gray', 'lavender','lightpink'] , explode=explode, shadow=True, startangle=90)
plt.legend(loc ="lower left")
```

```python
plt.xlabel("Amount of Close Friends")
plt.ylabel("Percent")
plt.title("How many close friends do you have?")
df['Close_Friends'].value_counts(normalize = True).plot(kind ='barh', color = 'maroon', width = 0.7, alpha = 0.5, grid = True)
```
